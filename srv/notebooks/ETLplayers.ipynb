```python
playerOverall = players[["long_name","club_name", "club_position","overall"]
].sort_values(by="overall", ascending=False)
playerOverall
```

```python
goalkeepers = playerOverall[playerOverall["club_position"]=="GK"]
goalkeepers.head()
```

```python
# check how many goalkeepers are there per team
goalkeepers.groupby("club_name").count()
```

```python
#get only the first goalkeeper
goalkeepers = goalkeepers.sort_values(by="overall", ascending=False).drop_duplicates(
    subset="club_name")
goalkeepers
```

```python
# get forwards
forward = playerOverall[playerOverall["club_position"].isin(["ST", "RW", "LW", "CF", 
                                                             "RCM","LCM","CAM","RDM","LCM","CM","CDM","LM","RM"])]
forward.head()
```

```python
#get only the first forward
forward = forward.sort_values(by="overall", ascending=False).drop_duplicates(
    subset="club_name")
forward
```

```python
# get mid field 1
mid1 = playerOverall[playerOverall["club_position"].isin(["RCM","LCM","CAM","RDM","LCM","CM","CDM","LM","RM"])]
mid1.head()
```

```python
# get only names not in forward
mid1 = mid1[~mid1.long_name.isin(forward.long_name.values)]
```

```python
#get only the first midfielder
mid1 = mid1.sort_values(by="overall", ascending=False).drop_duplicates(
    subset="club_name")
mid1.head()
```

```python
# get mid field 2
mid2 = playerOverall[playerOverall["club_position"].isin(["RCM","LCM","CAM","RDM","LCM","CM","CDM","LM","RM"])]
mid2.head()
```

```python
# get only names not in mid 1 and forward
mid2 = mid2[(~mid2.long_name.isin(mid1.long_name.values) & ~mid2.long_name.isin(forward.long_name.values))]
```

```python
#get only the next mid after first midfielder
mid2 = mid2.sort_values(by="overall", ascending=False).drop_duplicates(
    subset="club_name")
mid2.head()
```

```python
# get defender field 1
def1 = playerOverall[playerOverall["club_position"].isin(["LB","RB","RCB","LCB","CB","LWB","RWB"])]
def1.head()
```

```python
#get only the first defender
def1 = def1.sort_values(by="overall", ascending=False).drop_duplicates(
    subset="club_name")
def1
```

```python
# get defender field 2
def2 = playerOverall[playerOverall["club_position"].isin(["LB","RB","RCB","LCB","CB","LWB","RWB"])]
def2.head()
```

```python
# get only names not in def 1
def2 = def2[~def2.long_name.isin(def1.long_name.values)]
```

```python
#get only the next def after first defender
def2 = def2.sort_values(by="overall", ascending=False).drop_duplicates(
    subset="club_name")
def2
```

```python
#load previously selected EPL club names
selected = pd.read_pickle("selectedNames.pkl")
```

```python
# filter team players for only EPL clubs

teamPlayers = teamPlayers[teamPlayers.club_name.isin(selected["Current Club"].unique())]
teamPlayers
```

```python
# merge all team names set
teamPlayers = pd.merge(teamPlayers, goalkeepers[["long_name","club_name"]], how="left", on=["club_name"]
                      ).rename(columns={"long_name": "GK"})
teamPlayers
```

```python
# merge all team names set
teamPlayers = pd.merge(teamPlayers, forward[["long_name","club_name"]], how="left", on=["club_name"]
                      ).rename(columns={"long_name": "FW"})
teamPlayers
```

```python
# merge midfields
teamPlayers = pd.merge(teamPlayers, mid1[["long_name","club_name"]], how="left", on=["club_name"]
                      ).rename(columns={"long_name": "MD1"})
teamPlayers
```

```python
# merge midfields
teamPlayers = pd.merge(teamPlayers, mid2[["long_name","club_name"]], how="left", on=["club_name"]
                      ).rename(columns={"long_name": "MD2"})
teamPlayers
```

```python
# merge defenders
teamPlayers = pd.merge(teamPlayers, def1[["long_name","club_name"]], how="left", on=["club_name"]
                      ).rename(columns={"long_name": "DF1"})
teamPlayers
```

```python
# merge defenders
teamPlayers = pd.merge(teamPlayers, def2[["long_name","club_name"]], how="left", on=["club_name"]
                      ).rename(columns={"long_name": "DF2"})
teamPlayers
```
