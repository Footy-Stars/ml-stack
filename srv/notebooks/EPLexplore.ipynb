```python
teamPlayers = players[["full_name","Current Club"]].sort_values(by="Current Club").groupby("Current Club").count()
```

```python
teamPlayers.reset_index(inplace=True)
del teamPlayers["full_name"]
teamPlayers
```

```python
playerminutes = players[["full_name","Current Club", "position","minutes_played_overall"]
][players["minutes_played_overall"]>500].sort_values(by="minutes_played_overall", ascending=False)
playerminutes
```

```python
goalkeeper = playerminutes[playerminutes["position"]=="Goalkeeper"]
```

```python
# check how many goalkeepers are there per team
goalkeeper.groupby("Current Club").count()
```

```python
#get only the first goalkeeper
goalkeeper = goalkeeper.sort_values(by="minutes_played_overall", ascending=False).drop_duplicates(
    subset="Current Club")
goalkeeper
```

```python
forward = playerminutes[playerminutes["position"]=="Forward"]
forward
```

```python
#get only the first forward
forward = forward.sort_values(by="minutes_played_overall", ascending=False).drop_duplicates(
    subset="Current Club")
forward
```

```python
mid1 = playerminutes[playerminutes["position"]=="Midfielder"]
mid1
```

```python
#get only the first midfielder
mid1 = mid1.sort_values(by="minutes_played_overall", ascending=False).drop_duplicates(
    subset="Current Club")
mid1
```

```python
players[["full_name","position","Current Club", "minutes_played_overall"]][(players["Current Club"]=="Manchester United") & (players.position=="Midfielder")]
```

```python
mid2 = playerminutes[playerminutes["position"]=="Midfielder"]
mid2
```

```python
# get only names not in mid 1
mid2 = mid2[~mid2.full_name.isin(mid1.full_name.values)]
```

```python
#get only the next mid after first midfielder
mid2 = mid2.sort_values(by="minutes_played_overall", ascending=False).drop_duplicates(
    subset="Current Club")
mid2
```

```python
mid2
```

```python
def1 = playerminutes[playerminutes["position"]=="Defender"]
```

```python
#get only the first defender
def1 = def1.sort_values(by="minutes_played_overall", ascending=False).drop_duplicates(
    subset="Current Club")
def1
```

```python
def2 = playerminutes[playerminutes["position"]=="Defender"]
```

```python
# get only names not in def 1
def2 = def2[~def2.full_name.isin(def1.full_name.values)]
```

```python
#get only the next def after first defender
def2 = def2.sort_values(by="minutes_played_overall", ascending=False).drop_duplicates(
    subset="Current Club")
def2
```

```python
# merge all team names set
teamPlayers = pd.merge(teamPlayers, goalkeeper[["full_name","Current Club"]], how="left", on=["Current Club"]
                      ).rename(columns={"full_name": "gk"})
teamPlayers
```

```python
# merge all team names set
teamPlayers = pd.merge(teamPlayers, forward[["full_name","Current Club"]], how="left", on=["Current Club"]
                      ).rename(columns={"full_name": "fw"})
teamPlayers
```

```python
# merge midfields
teamPlayers = pd.merge(teamPlayers, mid1[["full_name","Current Club"]], how="left", on=["Current Club"]
                      ).rename(columns={"full_name": "md1"})
teamPlayers
```

```python
# merge midfields
teamPlayers = pd.merge(teamPlayers, mid2[["full_name","Current Club"]], how="left", on=["Current Club"]
                      ).rename(columns={"full_name": "md2"})
teamPlayers
```

```python
# merge defenders
teamPlayers = pd.merge(teamPlayers, def1[["full_name","Current Club"]], how="left", on=["Current Club"]
                      ).rename(columns={"full_name": "df1"})
teamPlayers
```

```python
# merge defenders
teamPlayers = pd.merge(teamPlayers, def2[["full_name","Current Club"]], how="left", on=["Current Club"]
                      ).rename(columns={"full_name": "df2"})
teamPlayers
```

```python
teamPlayers.to_pickle("teams.pkl")
```

```python
# get list of all player names
selectedNames = pd.concat([goalkeeper, forward, mid1, mid2, def1, def2])
```
